```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("https://slemeshevsky.github.io/python-course/pandas/html/src-pandas/la-crimes-sample.csv.zip")

print(data.shape)
```

```python
print(data.info())
```

```python
data_types = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
data.select_dtypes(include=data_types).describe()
```

```python
obj = data[['Victim Age', 'Victim Sex', 'Victim Descent']].value_counts()
sns.set(rc={'figure.figsize':(20,15)})
sns.countplot(data=obj.reset_index(), x='Victim Descent', hue='Victim Sex',palette='Pastel1')
```

```python
obj = data['Victim Descent'].value_counts()
obj.sort_values().plot(x='Victim Descent', kind="barh", color="r", figsize=(6,6));
```

```python
obj = data['Area Name'].value_counts()
obj.sort_values().plot(kind="barh", color='red', figsize=(10,10));
```

```python
data['Victim Age'].plot.kde();
```

```python

crimes = data[data['Area Name'].str.contains('LA')]['Crime Code Description'].value_counts()
crimes.head(10).sort_values().plot(kind="barh", figsize=(7,7));
```

```python
df_agg = data.groupby('Victim Age')['Victim Sex'].value_counts()
df_agg.unstack().plot(figsize=(10,10), colormap="bwr"); 
```

```python

df = data.groupby('Victim Sex')['Crime Code Description'].value_counts().unstack()

fem = df.loc['F'].sort_values(ascending=False).head(5)
males = df.loc['M'].sort_values(ascending=False).head(5)

df = pd.DataFrame({'F': fem, 'M': males})
df.plot(kind='barh', colormap='bwr', figsize=(8,8));
```

```python

df = data.groupby('Area Name')['Victim Descent'].value_counts().transform(lambda x: (x - x.min()) / (x.max() - x.min()))
df.unstack().plot(figsize=(30,30), kind='barh', colormap='tab10');
```
